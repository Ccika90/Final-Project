*****Testing the Hypothesis*****

```python
import scipy.stats as stats
from itertools import combinations
def compare_sales_channels(df, channel1='Online', channel2='In-store'):
    """
    Compare revenue between two sales channels using t-test.
    
    Args:
        df (pd.DataFrame): DataFrame containing sales data
        channel1 (str): First sales channel to compare (default: 'Online')
        channel2 (str): Second sales channel to compare (default: 'In-store')
        
    Returns:
        dict: Dictionary containing test results and statistics
    """
    # Calculate revenue if not already present
    if 'revenue' not in df.columns:
        df['revenue'] = (df['unit_price'] * df['quantity']) - df['discount']
    
    # Filter data for the two channels
    channel1_data = df[df['sales_channel'].str.lower() == channel1.lower()]['revenue']
    channel2_data = df[df['sales_channel'].str.lower() == channel2.lower()]['revenue']
    
    # Check if we have enough data
    if len(channel1_data) < 2 or len(channel2_data) < 2:
        return {"error": f"Insufficient data ({channel1}: {len(channel1_data)}, {channel2}: {len(channel2_data)})"}
    
    # Perform Welch's t-test (doesn't assume equal variances)
    t_stat, p_value = stats.ttest_ind(channel1_data, channel2_data, equal_var=False)
    
    # Calculate means and std
    mean1, mean2 = channel1_data.mean(), channel2_data.mean()
    std1, std2 = channel1_data.std(), channel2_data.std()
    
    # Determine results
    alpha = 0.05
    significant = p_value < alpha
    higher_revenue = channel1 if mean1 > mean2 else channel2
    
    # Visualization
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='sales_channel', y='revenue', 
                data=df[df['sales_channel'].str.lower().isin([channel1.lower(), channel2.lower()])])
    plt.title(f'Revenue Comparison: {channel1} vs {channel2}')
    plt.xlabel('Sales Channel')
    plt.ylabel('Revenue')
    plt.show()
    
    return {
        "channels": [channel1, channel2],
        "t_statistic": t_stat,
        "p_value": p_value,
        "mean_revenues": {channel1: mean1, channel2: mean2},
        "std_revenues": {channel1: std1, channel2: std2},
        "sample_sizes": {channel1: len(channel1_data), channel2: len(channel2_data)},
        "significant_difference": significant,
        "higher_revenue": higher_revenue,
        "interpretation": (
            f"There {'IS' if significant else 'IS NOT'} a significant difference (p={p_value:.4f}). "
            f"{higher_revenue} generates higher revenue on average."
        )
    }

# Example usage
if __name__ == "__main__":
   
    
    # Compare Online vs In-store channels
    results = compare_sales_channels(df, 'Online', 'In-store')
    
    # Print results
    print("\nSales Channel Revenue Comparison (T-Test)")
    print("="*60)
    print(f"Channels Compared: {results['channels'][0]} vs {results['channels'][1]}")
    print(f"\nMean Revenue:")
    print(f"- {results['channels'][0]}: ${results['mean_revenues'][results['channels'][0]]:.2f}")
    print(f"- {results['channels'][1]}: ${results['mean_revenues'][results['channels'][1]]:.2f}")
    print(f"\nStandard Deviation:")
    print(f"- {results['channels'][0]}: ${results['std_revenues'][results['channels'][0]]:.2f}")
    print(f"- {results['channels'][1]}: ${results['std_revenues'][results['channels'][1]]:.2f}")
    print(f"\nSample Sizes:")
    print(f"- {results['channels'][0]}: {results['sample_sizes'][results['channels'][0]]}")
    print(f"- {results['channels'][1]}: {results['sample_sizes'][results['channels'][1]]}")
    print(f"\nT-statistic: {results['t_statistic']:.4f}")
    print(f"P-value: {results['p_value']:.4f}")
    print(f"\nConclusion: {results['interpretation']}")
```

*****Discounts and Quantity Purchased*****

****Hypothesis 2: Higher discounts lead to an increase in the quantity of items purchased.****

```python
def test_discount_quantity_hypothesis(df, discount_threshold=0.2):
    """
    Test whether higher discounts lead to increased quantity purchased using t-test.
    
    Args:
        df (pd.DataFrame): DataFrame containing sales data
        discount_threshold (float): Threshold to split into high/low discount groups
        
    Returns:
        dict: Dictionary containing test results and statistics
    """
    # Data preparation
    if 'discount' not in df.columns or 'quantity' not in df.columns:
        return {"error": "Required columns ('discount' or 'quantity') not found"}
    
    # Create discount groups
    df['discount_group'] = np.where(df['discount'] >= discount_threshold, 
                                  'High Discount', 
                                  'Low Discount')
    
    # Split into groups
    high_discount = df[df['discount_group'] == 'High Discount']['quantity']
    low_discount = df[df['discount_group'] == 'Low Discount']['quantity']
    
    # Check sample sizes
    if len(high_discount) < 2 or len(low_discount) < 2:
        return {"error": f"Insufficient data (High: {len(high_discount)}, Low: {len(low_discount)})"}
    
    # Perform Welch's t-test (doesn't assume equal variances)
    t_stat, p_value = stats.ttest_ind(high_discount, low_discount, equal_var=False)
    
    # Calculate statistics
    high_mean = high_discount.mean()
    low_mean = low_discount.mean()
    effect_size = (high_mean - low_mean) / np.sqrt(
        (high_discount.var() + low_discount.var()) / 2
    )
    
    # Visualization
    plt.figure(figsize=(12, 6))
    
    plt.subplot(1, 2, 1)
    sns.boxplot(x='discount_group', y='quantity', data=df)
    plt.title('Quantity Purchased by Discount Group')
    
    plt.subplot(1, 2, 2)
    sns.scatterplot(x='discount', y='quantity', hue='discount_group', data=df)
    plt.axvline(x=discount_threshold, color='red', linestyle='--')
    plt.title('Discount vs Quantity Purchased')
    
    plt.tight_layout()
    plt.show()
    
    # Determine results
    alpha = 0.05
    significant = p_value < alpha
    supports_hypothesis = significant and (high_mean > low_mean)
    
    return {
        "discount_threshold": discount_threshold,
        "high_discount_mean": high_mean,
        "low_discount_mean": low_mean,
        "high_discount_count": len(high_discount),
        "low_discount_count": len(low_discount),
        "t_statistic": t_stat,
        "p_value": p_value,
        "effect_size": effect_size,
        "significant_difference": significant,
        "supports_hypothesis": supports_hypothesis,
        "interpretation": (
            f"The data {'supports' if supports_hypothesis else 'does not support'} "
            f"the hypothesis that higher discounts (â‰¥{discount_threshold:.0%}) lead to "
            f"higher quantities purchased (p={p_value:.4f}, effect size={effect_size:.2f}). "
            f"High discount mean: {high_mean:.2f}, Low discount mean: {low_mean:.2f}"
        )
    }

# Example usage
if __name__ == "__main__":
   
    
    # Test the hypothesis with default 20% discount threshold
    results = test_discount_quantity_hypothesis(df)
    
    # Print results
    print("\nDiscount vs Quantity Purchased Hypothesis Test")
    print("="*70)
    print(f"Discount Threshold: â‰¥{results['discount_threshold']:.0%}")
    print(f"\nHigh Discount Group (â‰¥{results['discount_threshold']:.0%}):")
    print(f"- Mean Quantity: {results['high_discount_mean']:.2f}")
    print(f"- Sample Size: {results['high_discount_count']}")
    print(f"\nLow Discount Group (<{results['discount_threshold']:.0%}):")
    print(f"- Mean Quantity: {results['low_discount_mean']:.2f}")
    print(f"- Sample Size: {results['low_discount_count']}")
    print(f"\nT-statistic: {results['t_statistic']:.4f}")
    print(f"P-value: {results['p_value']:.4f}")
    print(f"Effect Size (Cohen's d): {results['effect_size']:.2f}")
    print(f"\nConclusion: {results['interpretation']}")
```

*****Country and Average Order Value*****

****Hypothesis 3: Orders from certain countries have higher average order values.****

```python
from itertools import combinations
def test_country_order_values(df, top_n=10):
    """
    Perform pairwise t-tests to compare average order values among top countries.
    
    Args:
        df (pd.DataFrame): DataFrame containing sales data
        top_n (int): Number of top countries to compare
        
    Returns:
        dict: Dictionary containing test results and statistics
    """
    # Calculate order value if not already present
    if 'order_value' not in df.columns:
        df['order_value'] = df['unit_price'] * df['quantity']
    
    # Check required columns
    if 'country' not in df.columns:
        return {"error": "Country column not found in dataset"}
    
    # Clean country data
    df['country'] = df['country'].str.strip().str.title()
    
    # Get top countries by order count
    top_countries = df['country'].value_counts().head(top_n).index.tolist()
    if len(top_countries) < 2:
        return {"error": "Insufficient countries for comparison"}
    
    # Prepare results storage
    results = {
        'country_comparisons': [],
        'top_countries': top_countries,
        'country_stats': {}
    }
    
    # Calculate statistics for each country
    for country in top_countries:
        country_data = df[df['country'] == country]['order_value']
        results['country_stats'][country] = {
            'mean': country_data.mean(),
            'std': country_data.std(),
            'count': len(country_data)
        }
    
    # Perform pairwise comparisons
    for country1, country2 in combinations(top_countries, 2):
        data1 = df[df['country'] == country1]['order_value']
        data2 = df[df['country'] == country2]['order_value']
        
        if len(data1) < 2 or len(data2) < 2:
            continue  # Skip if insufficient data
            
        # Welch's t-test
        t_stat, p_value = stats.ttest_ind(data1, data2, equal_var=False)
        
        # Calculate effect size
        pooled_std = np.sqrt((data1.std()**2 + data2.std()**2) / 2)
        effect_size = (data1.mean() - data2.mean()) / pooled_std
        
        # Determine significance
        significant = p_value < 0.05
        higher_value = country1 if data1.mean() > data2.mean() else country2
        
        results['country_comparisons'].append({
            'countries': [country1, country2],
            't_statistic': t_stat,
            'p_value': p_value,
            'effect_size': effect_size,
            'significant': significant,
            'higher_value': higher_value
        })
    
    # Visualization
    plt.figure(figsize=(15, 8))
    
    # Boxplot for top countries
    plt.subplot(1, 2, 1)
    sns.boxplot(x='country', y='order_value', 
                data=df[df['country'].isin(top_countries)])
    plt.title(f'Order Value Distribution for Top {top_n} Countries')
    plt.xlabel('Country')
    plt.ylabel('Order Value')
    plt.xticks(rotation=45)
    
    # Bar plot of average order values
    plt.subplot(1, 2, 2)
    avg_values = df[df['country'].isin(top_countries)].groupby('country')['order_value'].mean()
    avg_values.sort_values(ascending=False).plot(kind='bar')
    plt.title(f'Average Order Value for Top {top_n} Countries')
    plt.ylabel('Average Order Value')
    plt.xticks(rotation=45)
    
    plt.tight_layout()
    plt.show()
    
    return results

def print_results(results):
    """Print the analysis results in a readable format"""
    if "error" in results:
        print(f"Error: {results['error']}")
        return
    
    print("\nCOUNTRY ORDER VALUE ANALYSIS")
    print("="*70)
    
    # Print country statistics
    print("\nCOUNTRY STATISTICS:")
    stats_df = pd.DataFrame(results['country_stats']).T
    stats_df['mean'] = stats_df['mean'].apply(lambda x: f"${x:,.2f}")
    stats_df['std'] = stats_df['std'].apply(lambda x: f"${x:,.2f}")
    print(stats_df)
    
    # Print significant comparisons
    print("\nSIGNIFICANT DIFFERENCES (p < 0.05):")
    sig_comps = [c for c in results['country_comparisons'] if c['significant']]
    if not sig_comps:
        print("No significant differences found")
    else:
        for comp in sig_comps:
            print(f"\n{comp['countries'][0]} vs {comp['countries'][1]}:")
            print(f"  Higher Value: {comp['higher_value']}")
            print(f"  T-statistic: {comp['t_statistic']:.2f}")
            print(f"  P-value: {comp['p_value']:.4f}")
            print(f"  Effect Size: {comp['effect_size']:.2f}")

# Example usage
if __name__ == "__main__":
   
    
    # Run the analysis
    results = test_country_order_values(df, top_n=10)
    
    # Print results
    print_results(results)
    
```

```python
from scipy.stats import ttest_ind
import itertools

# Assuming 'df' is your original DataFrame with 'category' and 'discount' columns
# Get unique categories
categories = df['category'].unique()

# Prepare a list to store results
results = []

# Perform pairwise t-tests between categories
for cat1, cat2 in itertools.combinations(categories, 2):
    # Extract discount data for each category
    discounts_cat1 = df[df['category'] == cat1]['discount']
    discounts_cat2 = df[df['category'] == cat2]['discount']
    
    # Perform two-sample t-test (Welch's t-test, does not assume equal variances)
    t_stat, p_value = ttest_ind(discounts_cat1, discounts_cat2, equal_var=False)
    
   
    #results['category_comparisons'].append({
            #'categories': [cat1, cat2],
            #'t_statistic': t_stat,
            #'p_value': p_value,
            #'effect_size': effect_size,
            #'significant': significant,
            #'higher_discount': higher_discount
        #})
    # Store the results
    results.append({
        'Category 1': cat1,
        'Category 2': cat2,
        'T-statistic': t_stat,
        'P-value': p_value,
       
    })

# Convert results to DataFrame for better readability
results_df = pd.DataFrame(results)

# Display the results
print(results_df)

# Optional: Interpret results at alpha = 0.05 significance level
alpha = 0.05
results_df['Significant Difference'] = results_df['P-value'] < alpha

print("\nInterpretation (Significant differences at alpha=0.05):")
print(results_df[['Category 1', 'Category 2', 'Significant Difference']])
```
